```python
Q = 'Age(year)'
df0 = df
df0['Age(year)']= df0['Age']
df0['Age(year)']= df0['Age(year)'].replace({1:'18-24', 2:'25-29', 3:'30-34', 4:'35-39', 5:'40-44', 6:'45-49'
                                            , 7:'50-54', 8:'55-59', 9:'60-64', 10:'65-69', 11:'70-74', 12:'75-79', 13:'80 or older'})

df0["Diabetes"] = df0["Diabetes"].replace({ 0 :"No_Diabetes", 1:"Diabetes", 2:"Diabetes"})
Di_L = []
age_L = []

for a in df0[Q].unique():
    df00 = df0[df0[Q]==a]
    p = len(df00[df00["Diabetes"]=="Diabetes"])/len(df00)
    p = 100 * round(p,2)
    Di_L.append(p)
    age_L.append(a)
dp= pd.DataFrame({Q: age_L , "% of Diabetes":Di_L})
dp.sort_values("% of Diabetes")
```
